```python
import random
```

```python
fichas = list(range(1, 11))

ntrials = 1000000
sucesso = 0

for i in range(ntrials):
  b1, b2 = random.sample(fichas, 2)

  sucesso += 1 if b1+b2 ==12 else 0

print(f"Probabilidade: {sucesso/ntrials:.3%}")
```
